```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os

%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly,plot_components_plotly
```

```python
df['y'] = df['화재총합(건)']
df['ds'] = df['연도']
df_new = df[['ds','y']]
df_new
```

### 이상치 제거 없이 예측

```python
m = Prophet()
m.fit(df_new)
```

```python
future = m.make_future_dataframe(periods=12,freq='M')
future.tail()
```

```python
forecast = m.predict(future)
forecast.tail()
```

```python
fig = m.plot(forecast)
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']].iloc[-13:]
```

```python
fig = m.plot_components(forecast)
```

### 교차검증

```python
m = Prophet()
m.fit(df_new)
forecast = m.predict(future)
```

```python
cutoffs = pd.to_datetime(['2022-12-31'])
```

```python
from prophet.diagnostics import cross_validation
#df_cv = cross_validation(m,initial='1460 days',period='30 days',horizon='30 days')
df_cv = cross_validation(m,cutoffs=cutoffs,horizon='365 days',parallel='processes')
df_cv.head()
```

```python
df_cv
```

## 모델평가

```python
from prophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p
```

```python
from prophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv,metric='mape')
```

### 하이퍼파라미터 튜닝

```python
cutoffs = pd.to_datetime(['2022-12-31'])
```

```python
import itertools
import numpy as np
import pandas as pd

param_grid = {
    'changepoint_prior_scale' : [0.001,0.01,0.1,0.3,0.5,0.6],
    'seasonality_prior_scale' : [0.01,0.1,1.0,2.0,3.0,5.0],
}

all_params = [dict(zip(param_grid.keys(),v)) for v in itertools.product(*param_grid.values())]
mapes = []

for params in all_params:
    m = Prophet(**params).fit(df_new)
    df_cv = cross_validation(m,cutoffs=cutoffs,horizon='365 days',parallel='processes')
    #df_cv = cross_validation(m,initial='1460 days',period='30 days',horizon='30 days',parallel='processes')
    df_p = performance_metrics(df_cv,rolling_window=1)
    mapes.append(df_p['mape'].values[0])

tuning_results = pd.DataFrame(all_params)
tuning_results['mape'] = mapes
print(tuning_results)
```

```python
best_params = all_params[np.argmin(mapes)]
best_params
```

```python
m = Prophet(changepoint_prior_scale=0.1,seasonality_prior_scale=0.1)
m.fit(df_new)
forecast = m.predict(future)
fig = m.plot(forecast)
```
